```python
import psycopg2 as ps
import pandas.io.sql as sqlio
```

```python
# Providing the postgres username and password 
connect = ps.connect(dbname = "data",
                        user = "AyushKatiyar",      
                        password = "Ayush@12345",
                        host = "database",
                        port = "5432")
```

```python
#establishing connection info
print(connect.get_dsn_parameters())
```

```python
# Select all details from table
df = sqlio.read_sql_query("""select * from user_details""", connect)
df.head(10)
```

```python
# Select all user_details who's name ends with john
df = sqlio.read_sql_query("""select * from user_details where last_name='john'""", connect)
df.head(10)
```

```python
# select all females 
df = sqlio.read_sql_query("""select * from user_details where gender='Female'""", connect)
df.head(10)
```

```python
#select all names like %da%
df = sqlio.read_sql_query("""select * from user_details where first_name LIKE '%da%'""", connect)
df.head(20)
```

```python
# Select first_name and its count for all first names in ascending order
df = sqlio.read_sql_query("""select first_name, COUNT(first_name) AS cnt from user_details GROUP BY first_name ORDER BY cnt ASC""", connect)
df.head(10)
```
